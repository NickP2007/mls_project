```python
# Import glob to read multiple CSV files into one df rather than separate dfs
import glob

# Get CSV files list from folder
path = r"C:\Users\nickp\Documents\Python Projects\mls_data"
csv_files = glob.glob(path + "/*.csv")
```

```python
# Read each CSV file into DataFrame
# Creating a list of dataframes
df_list = (pd.read_csv(file, low_memory=False) for file in csv_files)
```

```python
# Concat all dataframes
big_df = pd.concat(df_list, ignore_index=True)
```

```python
# Alternate approach with map function
def readcsv(args):
    return pd.read_csv(args, low_memory=False, header=None)

df = pd.concat(map(readcsv, csv_files))
```
